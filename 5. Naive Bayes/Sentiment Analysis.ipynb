**Cleaning Text-**
- remove html symbols
- remove special characters
- remove stopwords
- stemming

```python
import re

def remove_html_tags(text):
    clean_text = re.sub(r'<.*?>',' ',text)
    return clean_text
```

```python
data['review'] = data['review'].apply(remove_html_tags)
```

```python
data['review'][0]
```

```python
def convert_to_lower(text):
    return text.lower()
```

```python
data['review']=data['review'].apply(convert_to_lower)
```

```python
def remove_special_characters(text):

    x = " "

    for i in text:
        if i.isalnum():
            x=x+i
        else:
            x = x + " "
    return x
```

```python
data['review']=data['review'].apply(remove_special_characters)
```

```python
import nltk
from nltk.corpus import stopwords
```

```python
stopwords.words('english')
```

```python
def remove_stopwords(text):
    x = []

    for i in text.split():
        if i not in stopwords.words('english'):
            x.append(i)
    y = x[:]
    x.clear()
    return y
```

```python
data['review']=data['review'].apply(remove_stopwords)
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stemming(text):

    x=[]
    for i in text:
        x.append(ps.stem(i))

    y = x[:]
    x.clear()
    return y
```

```python
stemming(['play','playful', 'playing'])
```

```python
data['review']=data['review'].apply(stemming)
```

```python
def word_joiner(text):
    return " ".join(text)
```

```python
data['review']=data['review'].apply(word_joiner)
```
