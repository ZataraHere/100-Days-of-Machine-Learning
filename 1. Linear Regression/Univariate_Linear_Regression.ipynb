```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/ZataraHere/100-Days-of-Machine-Learning/refs/heads/main/Data/placement.csv'
```

```python
df = pd.read_csv(url)
```

```python
df.head()
```

```python
df.shape
```

```python
sns.histplot(x ='cgpa',kde = True,data = df)
```

```python
sns.histplot(x ='package',kde = True,data = df)
```

```python
sns.catplot(x= 'cgpa', kind = 'box', data = df)
```

```python
sns.catplot(x= 'package', kind = 'box', data = df)
```

```python
sns.relplot(x=df['cgpa'],y=df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
X = df.iloc[:,0:1]
y = df.iloc[:,-1]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train,y_train)
```

```python
X_test
```

```python
y_test
```

```python
X_test.shape
```

```python
lr.predict(X_test.iloc[0].values.reshape(1,1))
```

```python
plt.scatter(df['cgpa'],df['package'])
plt.plot(X_train,lr.predict(X_train),color='red')
plt.xlabel('CGPA')
plt.ylabel('Package(in lpa)')
```

```python
m = lr.coef_
m
```

```python
b = lr.intercept_
```

```python
# y = mx + b

m * 8.58 + b
```
